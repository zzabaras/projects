## 1. Установка Jupyter notebook

1) Самый простой способ установить все разом это [установить дистрибутив Anaconda](https://www.anaconda.com/download/) для своей операционной системы. В этом дистрибутиве уже есть  `jupyter notebook`, небольшая ide и предустановлен широкий спект пакетов.

_Наши шаги:_ <br>
* Скачать дистрибутив для своей операционной системы
* Открыть командную строку/терминал
* Запустить Jupyter notebook, написав в консоли `jupyter notebook`

Через несколько секунд должна открыться страница в браузере со списком файлов директории, в котороый была запущена команда (можно не ждать и самим перейти по http://localhost:8888/ - обычно там располагается локальный "сервер" Jupyter Notebook`

2) Альтернативный версии установки (например через `pip`) лежит [тут](http://jupyter.readthedocs.org/en/latest/install.html)

Чтобы убедиться, что у нас все установилось можно проверить версию можно напечатав в командной строке/консоле/терминале `python --version` и нажав `Enter`
